## 1. Datos de partida. Apertura de archivos.

A conitnuación se realiza la apertura de los archivos y su conversión en dataframes. Para obtener la estructura deseada en cada dataframe se ha realizado una pequeña limpieza preliminar de los datos durante el proceso

```python
# Abrimos el primer fichero de csv
df_csv_origen = pd.read_csv("../datos/exported_db_data.csv")
df_csv_origen.head(2)
```

```python
# creamos una copia llamada df_csv poniendo la columna index como index
df_csv = df_csv_origen.set_index(['index'], inplace=False)
```

```python
# comprobamos que todo esta correcto
df_csv.head(2)
```

```python
#abrimos el 2º archivo csv como kaggle
df_kaggle_origen = pd.read_csv("../datos/data_remaning_kaggle.csv", index_col= 0)
df_kaggle_origen.head(2)
```

```python
# creamos una copia llamada df_kaggle poniendo la columna index como index
df_kaggle = df_kaggle_origen.set_index(['index'], inplace=False)
```

```python
# comprobamos que todo esta correcto
df_kaggle.sample(3)
```

```python
# Abrimos el archivo .xml
tree = ET.parse('../datos/data_xml_clean.xml')
root = tree.getroot()


xml_list = []
for child in root:
    xml_dict = {}
    for subchild in child:
        xml_dict[subchild.tag] = subchild.text
    xml_list.append(xml_dict) # Es una lista que contiene diccionarios

df_xml_origen = pd.DataFrame.from_dict(xml_list)
```

```python
df_xml_origen.head(2)
```

```python
# creamos una copia (df_xml), eliminamos la columna 'level_0' por ser redundante y ponemos la columna 'index' como índice
df_xml = df_xml_origen.drop(['level_0'], axis=1)
df_xml.set_index(['index'], inplace= True)
```

```python
# mostramos el resultado
df_xml.head(3)
```

```python
# abrimos el archivo .txt
f = open("../datos/data_txt_clean.txt", mode='r', encoding='utf-8')
fichero_txt = f.readlines()
```

```python
# printeamos una muestra del resultado (una lista de strings)
fichero_txt[:4]
```

```python
# realizamos una pequeña limpieza de 
fichero_txt_lista = []
for i in fichero_txt:
    i = i.strip('\n') # eliminamos los \n
    i = i.split(';') # convertimos cada string en una lista de strings dividiendo por los ';'
    fichero_txt_lista.append(i)
```

```python
# printeamos una muestra del resultado (una lista de listas de strings)
fichero_txt_lista[:4]
```

```python
# creamos el dataframe del .txt
df_txt_origen = pd.DataFrame(fichero_txt_lista,columns= fichero_txt_lista[0])
df_txt_origen.head()
```

```python
# creamos una copia del dataframe original y sobre ella realizamos algunos cambios
df_txt = df_txt_origen.copy()
df_txt.set_index(['index'], inplace= True) # incluimos la columna 'index' como índice
df_txt.drop(['index'],axis= 0, inplace= True) # eliminamos la primera fila que corresponde al número de las preguntas
df_txt.head()
```

## 3. Unión de los datos en un único dataframe

```python
# unimos por el índice los datos procedentes del .xml y el .txt
df_uniones = df_xml.join(df_txt, on = "index")
df_uniones.head()
```

```python
# unimos por el índice los datos procedentes de ambos .csv (df_csv y df_kaggle)
df_uniones2 = df_csv.join(df_kaggle, on = "index")
df_uniones2.head()
```

```python
# dado que nos da error al unir los dos dataframes anteriores en uno solo usando .join
# lo resolvemos reseteando los índices y uniéndolos con pd.concat()
df_uniones.reset_index(drop = True, inplace= True)
```

```python
df_uniones2.reset_index(drop = True, inplace= True)
```

```python
df_union_final = pd.concat([df_uniones,df_uniones2], axis=1)
df_union_final.head(2)
```

```python
# guardamos en un .csv el dataframe completo
df_union_final.to_csv('../resultados/df_completo_sin_tratar.csv')
```
